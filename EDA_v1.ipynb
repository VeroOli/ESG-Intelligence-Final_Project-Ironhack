```python
df_semi=pd.read_csv('data/semi.csv')

df_semi
```

```python
df_semi.replace("#N/A Field Not Applicable", np.nan, inplace=True)
df_semi.replace("#N/A Authorization", np.nan, inplace=True)#N/A Authorization
```

```python
df_semi[df_semi.isnull().any(axis=1)]
```

```python
columns_with_nulls = df_semi.columns[df_semi.isnull().any()].tolist()
columns_with_nulls
```

```python
null_percentage = df_semi.isnull().mean() * 100
columns_with_nulls_percentage = null_percentage[null_percentage > 0].sort_values(ascending=False)
columns_with_nulls_percentage 
```

```python
columns_list = null_percentage[null_percentage <20].index.tolist()
len(columns_list)
(df_semi[columns_list])
```

```python
columns_list = null_percentage[null_percentage > 0].index.tolist()
len(columns_list)
```

```python
df_semi_clean= df_semi.dropna(axis=1)
df_semi_clean
```

```python
columns_with_only_zeros = df_semi_clean.columns[(df_semi_clean == 0).all()]
columns_with_only_zeros
```

```python
df_clean = df_semi_clean.drop(columns=columns_with_only_zeros)
df_clean 
```

```python
percentage_zeros = (df_clean.eq(0).sum() / len(df_clean)) * 100
columns_list = percentage_zeros[percentage_zeros > 20].index.tolist()
print(len(columns_list))
print(columns_list)
```

```python
df_clean.drop(columns=columns_list, inplace=True)
df_clean
```
